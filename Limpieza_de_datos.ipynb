###¿Cuál es la proporción de cuentas registradas que han realizado al menos un cobro o un pago mediante CoDi?

El dataframe df_una_transaccion contiene datos acumulativos, entonces para sacar la proporcion de cuentas validadas que han realizado por lo menos un pago o un cobro se tomara el ultimo dato del dataframe. 

Para sacar la proporción se dividirá el numero de cuentas registradas que han realizado por lo menos un cobro o un pago entre el total de cuentas validadas.

EL porcentaje calculado fue: 
* Cuentas registradas con al menos un pago: 9.3 %
* Cuentas registradas con al menos un cobro: 6.2 %

```python

#El dataframe una transaccion es acumulativo , entonces para sacar la proporcion de cuentas validadas 
#que han realizado por lo menos un pago o un cobro se tomara el ultimo dato

df_una_transaccion.iloc[-1,1]  #Cuentas que han realizado por lo menos un pago
df_una_transaccion.iloc[-1,2]  #Cuentas que han realizado por lo menos un cobro
df_cuentas_validas.sum().head(1) # Total de cuentas validadas

proporción_cuentas_un_pago = df_una_transaccion.iloc[-1,1]/df_cuentas_validas.sum().head(1) 
proporción_cuentas_un_cobro = df_una_transaccion.iloc[-1,2]/df_cuentas_validas.sum().head(1)  

print(f"El porcentaje de cuentas que han realizado por lo menos un pago es : {round(proporción_cuentas_un_pago*100,2)}")
print(f"El porcentaje de cuentas que han realizado por lo menos un cobro es : {round(proporción_cuentas_un_cobro*100,2)}")

```

### ¿Cúal ha sido el promedio por año de las cuentas que han realizado al menos un cobro o un pago mediante CoDi?

El dataframe una transaccion es acumulativo , para sacar los datos por fecha sacariamos los datos que corresponden al ultimo día del año , dividido entre el numero de días (365)

Para el caso del año 2019 , tenemos datos a partir del 23 de Octubre , el número de días al 31 de diciembre son 69 , por lo que en lugar de dividir entre 365 días , se dividirá entre 69 días

* 31/12/2019
* 31/12/2020
* 31/12/2021
* 31/12/2022

Promedio de cuentas en 2019 han realizado por lo menos un cobro o un pago: 

* Por lo menos un pago: 1030.84 transacciones diarias
* Por lo menos un cobro: 878.3 transacciones diarias

Promedio de cuentas en 2020 han realizado por lo menos un cobro o un pago: 

* Por lo menos un pago: 721.48 transacciones diarias
* Por lo menos un cobro: 585.26 transacciones diarias

Promedio de cuentas en 2021 han realizado por lo menos un cobro o un pago: 

* Por lo menos un pago: 1190.53  transacciones diarias
* Por lo menos un cobro: 935.76 transacciones diarias

Promedio de cuentas en 2022 han realizado por lo menos un cobro o un pago: 

* Por lo menos un pago: 1210.38 transacciones diarias

###¿Cuál es el porcentaje de dispositivos Android y iOS registrados en CoDi?

Para resolver está pregunta sacamos un subconjunto del dataframe df_smarphones, ya que solo nos interesa el total de dispositivos y no el nuúmero de dispositivos por entidad financiera. Despues agregamos una columna para sumar el total de dispositivos enrolados. Con esas columnas tuvimos la información necesaria para sacar el porcentaje por sistema operativo.

El número de dispositivos enrolados de cada sistema operativo , cambia todos los días , pero en promedio tenemos que 74.92 % de los dispositivos que usan CoDi son android y el 25.08 % de los dispositivos unan ios.

```python
pd.options.mode.chained_assignment = None 
#Vamos a sacar el subconjunto de datos donde solo tengamos las columnas del total dispositivos androi y total dispositivos Ios
df_smarphones_android_ios_enrolados =df_smartphones[["total_de_dispositivos_android_enrolados","total_de_dispositivos_ios_enrolados"]]

#Agregamos una tercera columna que nos sume el total de dispositivos enrolados

df_smarphones_android_ios_enrolados["total_dispositivos"] = df_smarphones_android_ios_enrolados.apply(lambda x: x['total_de_dispositivos_android_enrolados'] + x['total_de_dispositivos_ios_enrolados'], axis=1)


#Sacamos el porcentaje de dispositivos android enrolados

df_smarphones_android_ios_enrolados["porcentaje_android"] = df_smarphones_android_ios_enrolados.apply(lambda x: round((x['total_de_dispositivos_android_enrolados'] / x['total_dispositivos'])*100,2), axis=1)

#Sacamos el porcentaje de dispositivos ios enrolados

df_smarphones_android_ios_enrolados["porcentaje_ios"] = df_smarphones_android_ios_enrolados.apply(lambda x: round((x['total_de_dispositivos_ios_enrolados'] / x['total_dispositivos'])*100,2), axis=1)

porcentaje_android = round(df_smarphones_android_ios_enrolados["porcentaje_android"].mean(),2)
porcentaje_ios =round(df_smarphones_android_ios_enrolados["porcentaje_ios"].mean(),2)

print("sistema_operativo".ljust(20),"porcentaje".ljust(10))
print("-"*20,"-"*10)
print("android".ljust(20),f"{porcentaje_android}")
print("ios".ljust(20),f"{porcentaje_ios}")
```
